**1.** Return uncommon elements of lists. Order of elements does not matter.

input:
    list1 = [1, 1, 2]
    list2 = [2, 3, 4]
output: [1, 1, 3, 4]

input:
    list1 = [1, 2, 3]
    list2 = [4, 5, 6]
output: [1, 2, 3, 4, 5, 6]

input:
    list1 = [1, 1, 2, 3, 4, 2]
    list2 = [1, 3, 4, 5]
output: [2, 2, 5]


```python
def uncommon(x, y):
    z = [i for i in x if i not in y]
    v = [i for i in y if i not in x]
    print(z+v)
    
list1 = [1, 1, 2, 3, 4, 2]
list2 = [1, 3, 4, 5]
uncommon(list1, list2)
```

**6. Prime Numbers**
Task: Write a Python program that prints all prime numbers between 1 and 100.

```python
primes = []
for i in range(2, 100):
    if not any(i%n == 0 for n in range(2, i)):
        primes.append(i)
print(primes)
```
