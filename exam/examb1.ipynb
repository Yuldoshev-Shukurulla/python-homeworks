Create a DataFrame from the data below (note the intentional issues: spaces in names, negative ages, missing salaries):
•  Clean names (strip whitespace, title case)
•  Replace negative ages with NaN, then fill all NaN ages with the department-wise median age
•  Fill NaN salaries with the overall mean salary
•  Add a column 'Salary_Rank' with each employee's rank within their department (highest salary = rank

data = {
  "Name": ["  alice", "BOB  ", " carol ", "dave", "  EVE"],
  "Age": [30, -5, 28, 45, 35],
  "Salary": [70000, 85000, None, 60000, None],
  "Department": ["IT", "IT", "HR", "HR", "IT"]

```python
import pandas as pd
import numpy as np
data = {
  "Name": ["  alice", "BOB  ", " carol ", "dave", "  EVE"],
  "Age": [30, -5, 28, 45, 35],
  "Salary": [70000, 85000, None, 60000, None],
  "Department": ["IT", "IT", "HR", "HR", "IT"]
}
df = pd.DataFrame(data)
df['Name'] = df['Name'].str.strip()
df['Name'] = df['Name'].str.title()
df['Age'] = np.where(df['Age'] > 0, df['Age'], np.nan)
df['Age'] = df['Age'].fillna(df.groupby('Department')['Age'].transform('mean'))
df['Salary'] = df['Salary'].fillna(round(df['Salary'].agg('mean'), 2))
df['Salary_rank'] = df.groupby('Department')['Salary'].rank(ascending=False, method='min')

df
```
