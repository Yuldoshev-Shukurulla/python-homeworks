Class Hierarchy with Operators  [4 pts]
Create a  Matrix  class that wraps a 2D list. Support:
•  __init__(self, data) — accepts a list of lists; validate all rows are equal length
•  __add__ — matrix addition (raise ValueError if dimensions differ)
•  __mul__ — scalar multiplication (int/float on right side)
•  transpose() — returns a new Matrix with rows/columns swapped
•  __str__ — display rows on separate lines
•  A class method  zeros(rows, cols)  that creates a zero-filled Matrix


```python
class Matrix():
    def __init__(self, data):
        first_row_len = len(data[0])
        if not all(len(row) == first_row_len for row in data):
            raise ValueError('All rows should be in same length.')
        self.data = data
    def __add__(self, other):
            if not isinstance(other, Matrix):
                raise TypeError("Faqat ikkita Matrix obyektini bir-biriga qo'shish mumkin!")

            if len(self.data) != len(other.data) or len(self.data[0]) != len(other.data[0]):
                raise ValueError('Matritsalar o\'lchami bir xil bo\'lishi shart!')

            new_data = []
            for i in range(len(self.data)):  # len(data) emas, len(self.data)
                row = []
                for k in range(len(self.data[i])):
                    row.append(self.data[i][k] + other.data[i][k])
                new_data.append(row)

            return Matrix(new_data)
    def __mul__(self, other):
            if not isinstance(other, Matrix):
                raise TypeError("Faqat ikkita Matrix obyektini bir-biriga qo'shish mumkin!")

            if len(self.data) != len(other.data) or len(self.data[0]) != len(other.data[0]):
                raise ValueError('Matritsalar o\'lchami bir xil bo\'lishi shart!')

            new_data = []
            for i in range(len(self.data)):  # len(data) emas, len(self.data)
                row = []
                for k in range(len(self.data[i])):
                    row.append(self.data[i][k] * other.data[i][k])
                new_data.append(row)

            return Matrix(new_data)
    def transpose(self):  
        new_data = [list(row) for row in zip(*self.data)]
        return Matrix(new_data)
    def __str__(self):
         return ',\n'.join(str(row) for row in self.data) 

@classmethod
def zeros(cls, rows, cols):
    zero_data = [[0 for _ in range(cols)] for _ in range(rows)]
    return cls(zero_data)    

a = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
b = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])       
    
```

  {"name": "  ALICE ", "age": "25", "salary": "50000"},

```python
records = [
  {"name": "  ALICE ", "age": "25", "salary": "50000"},
  {"name": "bob", "age": "thirty", "salary": "60000"},
  {"name": " Carol  ", "age": "35", "salary": ""}
]
def clean_records(records):
    cleaned = []
    for record in records:
        cleaned_record = {}
        if 'name' in record and record['name']:
            cleaned_record['name'] = record['name'].strip().title()
        else:
            cleaned_record['name'] = None

        try:
            cleaned_record['age'] = int(record['age'])
        except (ValueError, TypeError):
            cleaned_record['age'] = None
        
        try:
            cleaned_record['salary'] = int(record['salary'])
        except (ValueError, TypeError):
            cleaned_record['salary'] = None
        cleaned.append(cleaned_record)
    
    return cleaned
clean_records(records)
```

Grouping with Dicts  [4 pts]
Write  group_by_key(data, key)  that groups a list of dicts by a given key.
data = [
  {"dept": "IT", "name": "Alice"},
  {"dept": "HR", "name": "Bob"},
  {"dept": "IT", "name": "Carol"},
]
group_by_key(data, "dept") →
{"IT": [{"dept":"IT","name":"Alice"}, {"dept":"IT","name":"Carol"}],
 "HR": [{"dept":"HR","name":"Bob"}]}


```python
def group_by_key(data, key):
    result = {}
    for dict in data:
        if dict[key] not in result:
            result[dict[key]] = []
        result[dict[key]].append(dict)
    return result
data = [
  {"dept": "IT", "name": "Alice"},
  {"dept": "HR", "name": "Bob"},
  {"dept": "IT", "name": "Carol"},
]
print(group_by_key(data, 'dept'))
```

Create a  DataRow  class:
•  __init__(self, **kwargs)  — stores any key-value pairs as attributes
•  to_dict()  — returns all attributes as a dict
•  __getitem__(key)  — allows  row['name']  syntax
•  __repr__  — returns  DataRow(key=val, ...)


```python
class DataRow:
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)
    def to_dict(self):
        return self.__dict__
    def __getitem__(self, key):
        return self.__dict__.get(key)
    def __repr__(self):
        items = ", ".join(f"{key}={value}" for key, value in self.__dict__.items())
        return f"DataRow({items})"  
```
