Write a function  filter_by_frequency(lst, min_freq)  that returns a list containing only elements that appear at least  min_freq  times in  lst , maintaining their original order (first occurrence only). Handle empty lists.

```python
def filter_by_frequency(lst, min_freq):
    frequency = {}
    min_dict = {}
    result = []
    if lst:
        for i in lst:
            frequency[i] = frequency.setdefault(i, 0) +1
    else:
        print('List is empty')
    for k, v in frequency.items():
        if v >= min_freq:
            min_dict[k] = v
        else: continue
    for i in lst:
        if i in min_dict.keys() and i not in result:
            result.append(i)
        else: continue
    return result
a = [0, 1, 2, 3, 4, 5, 1, 4, 2, 3, 4, 2, 3, 6, 2]
filter_by_frequency(a, 3)
```

Given a list of tuples  records = [(name, subject, score), ...]  representing student scores, write a function  top_students(records, n)  that returns the top  n  students by average score across all their subjects. Return as a list of (name, avg_score) tuples sorted by avg_score descending. Use dictionaries for intermediate grouping.

```python

records = [
    ("Alice", "Math", 85),
    ("Bob", "Math", 90),
    ("Alice", "Science", 92),
    ("Bob", "Science", 55),
    ("Carol", "Math", 72),
    ("Carol", "Science", 80),
    ("Alice", "English", 78),
    ("Bob", "English", 88),
    ("Carol", "English", 95)
]
def top_students(data, n):
    scores = {}
    result = {}
    for row in data:
        scores[row[0]] = [0, 0]
    for row in data:
        scores[row[0]][0] += row[2]
        scores[row[0]][1] += 1
    for k, v in scores.items():
        result[k] = round(v[0]/v[1], 2)
    result_sorted = sorted(result.items(), key = lambda x: x[1], reverse=True)
    return result_sorted[:n]

top_students(records, 3)    

```

Create a class hierarchy for a media library:
•  Base: MediaItem(title, year) with __str__ and age() method returning current_year - year
•  Book(title, year, author, pages) — adds a reading_time() method (assume 2 min/page)
•  Movie(title, year, director, runtime_min) — adds is_long() returning True if > 120 min
•  Podcast(title, year, host, episodes) — adds total_hours(avg_min) method
•  Write a function  library_stats(items)  that takes a mixed list and prints count per type and average age


```python

class MediaItem():
    def __init__(self, title, year):
        self.title = title
        self.year = year
    def age(self):
        return 2026 - self.year

class Book(MediaItem):
    def __init__(self, title, year, author, pages):
        super().__init__(title, year)
        self.author = author
        self.pages = pages
    def reading_time(self):
        return self.pages * 2
    
class Movie(MediaItem):
    def __init__(self, title, year, director, runtime_min):
        super().__init__(title, year)
        self.director = director
        self.runtime_min = runtime_min
    def is_long(self):
        return True if self.runtime_min > 120 else False

class Podcast(MediaItem):
    def __init__(self, title, year, host, episodes):
        super().__init__(title, year)
        self.host = host
        self.episodes = episodes
    def total_hours(self, avg_min):
        return (self.episodes * avg_min) / 60
    
def library_stats(items):
    if not items:
        print('Library is empty!')
        return
    total_age = sum(i.age() for i in items)
    
    sorted_lib = {'Book' : 0, 'Movie' : 0, 'Podcast' : 0}
    for i in items:
        i_type = type(i).__name__
        if i_type in sorted_lib:
            sorted_lib[i_type] += 1

    avg_age = total_age / len(library)

    print(f"Counts per type: {sorted_lib}")
    print(f"Average age: {avg_age:.1f} years")
        

library = [
    Book("Atomic Habits", 2018, "James Clear", 320),
    Movie("Inception", 2010, "Christopher Nolan", 148),
    Movie("Toy Story", 1995, "John Lasseter", 81),
    Podcast("Huberman Lab", 2021, "Andrew Huberman", 150)
]

library_stats(library)
```
