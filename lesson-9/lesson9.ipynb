```python
from datetime import datetime
import json
class DateEncode(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, datetime):
            return obj.isoformat()
        return super().default(obj)

data = {
    'name' : 'John',
    'time' : datetime.now()
}

json.dumps(data, cls=DateEncode)
```
